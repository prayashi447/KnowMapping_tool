```python
yearly_avg = df.groupby("Year")["Total_Score"].mean()

plt.plot(yearly_avg.index, yearly_avg.values, marker="o")
plt.title("Average Total Score Over Years")
plt.xlabel("Year")
plt.ylabel("Average Total Score")
plt.show()
```

```python
grouped = df.groupby(["Year", "Department"])["Total_Score"].mean().unstack()

grouped.plot(kind="bar")
plt.title("Average Total Score by Year & Department")
plt.ylabel("Average Score")
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.title("Sine")

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.title("Cosine")

plt.subplot(2,2,3)
plt.plot(x, np.tan(x))
plt.title("Tangent")

plt.subplot(2,2,4)
plt.plot(x, np.exp(-x))
plt.title("Exponential")

plt.tight_layout()
plt.show()
```
